```python
for df in rock_island:
    for index, row in df.iterrows():
        if (not('Rock Island') in row['Admin2']):
            df.drop(index, inplace = True)
        elif ('Unassigned') in row['Admin2']:
            df.drop(index, inplace = True)
        elif ('Out of IL') in row['Admin2']:
            df.drop(index, inplace = True)
df_9_30   
```
